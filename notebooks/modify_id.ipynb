def add_unique_suffix(df, column):
    duplicates = df[df.duplicated(subset=[column], keep=False)]
    suffix_map = {}
    for value in duplicates[column].unique():
        suffix = 1
        for index, row in duplicates.iterrows():
            if row[column] == value:
                df.at[index, column] = f"{value}_{suffix}"
                suffix += 1
                suffix_map[value] = suffix
                return df

```python
def add_unique_suffix_2(df, column_name):
    # Track the counts of each value in the ID column
    value_counts = df[column_name].value_counts()

    # Iterate over the DataFrame rows
    for index, row in df.iterrows():
        id_value = row[column_name]
        count = value_counts[id_value]

        # If there is more than one occurrence of the ID value, add a suffix
        if count > 1:
            suffix = f'_{count - 1}'
            df.at[index, column_name] = f'{id_value}{suffix}'

        # Update the count after adding the suffix
        value_counts[id_value] -= 1

```

```python

# Call the function to add unique suffixes to the duplicate ID values
add_unique_suffix_2(df, 'SK_ID_CURR')
```
